Una característica sorprendente de Folium es la posibilidad de crear mapas de calor animados, cambiando los datos que se muestran en función de una determinada dimensión (por ejemplo, horas). Podemos lograrlo fácilmente usando el método de clase HeatMapWithTime (). 


Primero, creamos una lista anidada donde cada posición contiene la latitud y longitud de todos los accidentes automovilísticos en esa hora específica. Por ejemplo, hour_list [0] contiene los accidentes automovilísticos que ocurren desde las 00:00:00 hasta las 00:59:00 (por ejemplo, hour_list [0] → [[lat1, lon1], [lat2, lon2], [lat3, lon3] ,…, [Latn, logn]). Luego, llamamos al método y lo agregamos a nuestro mapa.

```python
from folium.plugins import HeatMapWithTime

# Crea un objeto mapa de la ciudad de Barcelona
barcelona_map= folium.Map(location=[df_traccar['Latitude'].mean(), df_traccar['Longitude'].mean()], zoom_start=12)

# Lista anidada que contiene la latitud y la longitud de los diferentes accidentes.
hour_list = [[] for _ in range(24)]
for lat,log,hour in zip(df_traccar.Latitude,df_traccar.Longitude,df_traccar.Date.dt.hour):
    hour_list[hour].append([lat,log]) 

# Etiquetas indicando las horas
index = [str(i)+' Horas' for i in range(24)]

# Cree una instancia de un mapa de calor con objeto de tiempo para los accidentes automovilísticos
HeatMapWithTime(hour_list, index).add_to(barcelona_map)

barcelona_map
```

```python
import folium

import altair as alt
import pandas as pd

# define data for demonstration
source = pd.DataFrame(
    {
        'a': ['1', '3', '6', '9', '12', '3', '6', '9', '12'],
        'b': [28, 55, 43, 91, 81, 53, 19, 87, 52],
    }
)

# create an altair chart, then convert to JSON
chart = alt.Chart(source).mark_bar().encode(x='a', y='b')
vis1 = chart.to_json()

# create folium map
start_coords = [df_traccar['Latitude'].mean(), df_traccar['Longitude'].mean()]
folium_map = folium.Map(
    location=start_coords,
    zoom_start=13,
    width='80%',
    height='80%',
)

# create a marker, with altair graphic as popup
circ_mkr = folium.CircleMarker(
    location=start_coords,
    radius=20,
    color='red',
    fill=True,
    fill_color='red',
    fillOpacity=1.0,
    opacity=1.0,
    tooltip='Distribuciones de Horas',
    popup=folium.Popup(max_width=400).add_child(folium.VegaLite(vis1, width=400, height=300)),
)

# add to map
circ_mkr.add_to(folium_map)

folium_map
```
